```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

# Example usage

ua = UserAgent()
headers = {"User-Agent": ua.random}
jobs = []
query = "computational biology"
location = "Switzerland"
pages = 2  # Number of pages to scrape

for page in range(pages):
    start = page * 25
    url = f"https://www.linkedin.com/jobs/search?keywords={query}&location={location}&start={start}"
    print(url)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    for job_card in soup.find_all("div", class_="base-card"):
        title = job_card.find("h3", class_="base-search-card__title")
        company = job_card.find("h4", class_="base-search-card__subtitle")
        link = job_card.find("a", class_="base-card__full-link")

        jobs.append({
            "title": title.get_text(strip=True) if title else None,
            "company": company.get_text(strip=True) if company else None,
            "url": link["href"] if link else None
        })


for job in jobs:
    print(f"{job['title']} â€” {job['company']}")
    print(f"ðŸ”— {job['url']}\n")
```

# Novartis


```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

def scrape_jobs_novartis(urls, keywords):
    results = []

    for url in urls:
        try:
            time.sleep(2)  # Respectful delay
            domain = urlparse(url).netloc
            headers = {'User-Agent': 'Mozilla/5.0'}
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')

            print(f"\nScraping {domain}...")
            
            if 'roche' in url:
                breakpoint()

            jobs_data = []
            job_links = []

            # Find job table rows
            for row in soup.select("table.views-table tbody tr"):
                try:
                    title_cell = row.find("td", class_="views-field-field-job-title")
                    business_cell = row.find("td", class_="views-field-field-job-business-unit")
                    country_cell = row.find("td", class_="views-field-field-job-country")
                    site_cell = row.find("td", class_="views-field-field-job-work-location")
                    date_cell = row.find("td", class_="views-field-field-job-posted-date")

                    job_title = title_cell.get_text(strip=True) if title_cell else None
                    job_link = urljoin(url, title_cell.find("a")["href"]) if title_cell and title_cell.find("a") else None
                    business = business_cell.get_text(strip=True) if business_cell else None
                    country = country_cell.get_text(strip=True) if country_cell else None
                    site = site_cell.get_text(strip=True) if site_cell else None
                    date_posted = date_cell.get_text(strip=True) if date_cell else None

                    if job_link:
                        job_links.append(job_link)

                    jobs_data.append({
                        "Job Title": job_title,
                        "Job Link": job_link,
                        "Business Unit": business,
                        "Country": country,
                        "Site": site,
                        "Date Posted": date_posted
                    })
                except Exception as inner_e:
                    print(f"Error parsing row: {inner_e}")

            # Keyword matching in page text
            text_content = soup.get_text().lower()
            keyword_hits = {kw: text_content.count(kw.lower()) for kw in keywords if kw.lower() in text_content}

            results.append({
                'Website': domain,
                'Jobs Found': len(jobs_data),
                'Job Links': list(set(job_links)),   # <-- flat list for compatibility
                'Jobs': jobs_data,                   # <-- structured details
                'Keyword Matches': keyword_hits,
                'Source URL': url
            })

        except Exception as e:
            print(f"Error processing {url}: {str(e)}")

    return pd.DataFrame(results)



# Example usage
urls = [
    'https://www.novartis.com/careers/career-search?search_api_fulltext=computational+biology&country%5B%5D=LOC_CH',
    # 'https://careers.roche.com/global/en/search-results?keywords=%22computational%20biology%22'
]

keywords = ['python', 'data science', 'bioinformatics', 'machine learning', 'computational biology']
df = scrape_jobs_novartis(urls, keywords)

# Display summary
for _, row in df.iterrows():
    print(f"\nðŸ” {row['Website']}: {row['Jobs Found']} jobs found")
    for link in row['Job Links']:
        print(f"â†’ {link}")
```
